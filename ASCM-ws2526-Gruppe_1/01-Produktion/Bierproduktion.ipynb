| Sorte  | DB je Liter|
|----------| ------|
| A       |  2,10 € |
| B       |  1,70 € |
| C       |  2,00 € |
| D       |  1,50 € |
| E       |  1,85 € |


In der nachstehenden Tabelle sind sowohl die Abfülldauer (AD) als auch die Rohstoffverbräuche (RV) für einen Liter der jeweiligen Biersorten ausgewiesen:

| Sorte  | AD | RV1 | RV2 | RV3 |
